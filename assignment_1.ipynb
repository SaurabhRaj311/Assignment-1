#Question 5: Given:
price = 499
discount = 0.15
final_price = price - price * discount
1. What is the data type of each variable?
2. Modify the code so that final_price is printed with 2 decimal places.
3. Convert price into a string and concatenate it into:
"The final price is ₹___"

**Answers ->**
Data Type in each variable:-
 - price -> int(integer)
 - discount -> float(decimal)
 - final_price -> float (result include decimal calculation)








```python
### Question 5 - CODE IMPLEMENTATION

price = 499          # integer
discount = 0.15      # float

# Calculate final price
final_price = price - price * discount

# Print with 2 decimal places
print(f"{final_price:.2f}")

# Convert to string and print
print("The final price is ₹ " + str(final_price))
```

#Question 7: You are given a customer review:
review = "The product quality is Excellent but delivery was Late"
Write Python code to:
1. Convert to lowercase
2. Count number of words
3. Check if the word "excellent" exists
4. Replace "Late" with "delayed"


**ANSWER ->**
- lower() -> converts text to lower case
- split() -> split sentence to words
- in -> will check existence of word
- replace() -> replace specific word
- len() -> count total words


```python
### Qestion 7 - CODE IMPLEMENTATION


review = "The product quality is Excellent but delivery was Late"

# 1. Convert to lowercase
lower_review = review.lower()

# 2. Count number of words
word_count = len(review.split())

# 3. Check if the word "excellent" exists
check_word = "excellent" in lower_review

# 4. Replace "Late" with "delayed"
updated_review = review.replace("Late", "delayed")

# Print results
print("Lowercase Review:", lower_review)
print("Word Count:", word_count)
print("Contains 'excellent':", check_word)
print("Updated Review:", updated_review)
```

#Question 8: Given sales data:
sales = [12000, 15000, 11000, 18000, 17000]
1. Calculate total revenue.
2. Find average revenue.
3. Identify highest sales month.
4. Append new month sale = 20000.
5. Sort the list in descending order.


**ANSWERS ->**
- sum() - make calculation of total
- len() - counts number of element
- max() - finds maximum value
- append() - add new value
- sort(reverse=true) - sort in descending order

```python
### Question 8 - CODE IMPLEMENTATION

sales = [12000, 15000, 11000, 18000, 17000]

# 1. Calculate total revenue
total_revenue = sum(sales)

# 2. Find average revenue
average_revenue = total_revenue / len(sales)

# 3. Identify highest sales month
highest_sales  = max(sales)
highest_month = sales.index(highest_sales) + 1  # +1 for month number

# 4. Append new month sale = 20000
sales.append(20000)

# 5. Sort the list in descending order
sales.sort(reverse=True)

# Print results
print("Total Revenue:", total_revenue)
print("Average Revenue:", average_revenue)
print("Highest Sales:", highest_sales)
print("Month of Highest Sales:", highest_month)
print("Updated Sales List:", sales)
```

#Question 9:You are building a simple sales analysis system.
Given:
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

Tasks:
1. Compute total revenue for each product.
2. Identify which product generated maximum revenue.
3. Store revenues in a new list.
4. Print a formatted summary like:

**ANSWERS ->**

- lists ( product, price and quantities)
- (for) for iteration
- append() -> for storing revenue
- max () - to find maximum value
- index() - to match product
- f-strings for the clean output

```python
### Question 9 - CODE IMPLEMENTATION

products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

# 1 & 3. Compute and store total revenue for each product
revenues = []
for i in range(len(products)):
    revenue = prices[i] * quantities[i]
    revenues.append(revenue)

# 2. Identify which product generated maximum revenue
max_revenue = max(revenues)
max_product = products[revenues.index(max_revenue)]

# Print formatted summary
print("Sales Summary:")
for i in range(len(products)):
  print(f"{products[i]} → Revenue: ₹{revenues[i]}")

print("\nProduct with Highest Revenue:", max_product)
```
